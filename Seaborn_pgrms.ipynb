# Distribution charts

```python
import seaborn as sns
import numpy as np
import pandas as pd

df=pd.read_excel("c:\\Datasets\\Superstore_sales.xls")
```

```python
sns.distplot(df.Sales)
# The output is a right skewed distribution
```

```python
sns.distplot(df.Profit)
```

```python
sns.distplot(df.Profit,kde=False)
```

```python
sns.distplot(df.Profit,hist=False)
```

```python
sns.distplot(df.Profit,kde=False,rug=True)
```

```python
sns.distplot(df.Profit,kde=False,rug=True,bins=100)
```

```python
sns.distplot(df.Profit[(df.Region=="West") & (df.Profit>100)],kde=False,rug=True,bins=100)
```
